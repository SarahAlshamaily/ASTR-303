```python
#Importing required packages

%pylab nbagg
%matplotlib inline
import numpy as np
```

```python
#Introducing variables

x = np.array([1])
y = np.array([0])
vx = np.array([0])
vy = np.array([1])
t1 = np.array([0])
```

```python
# Using the Euler method

f = 10 * 2 * np.pi
dt = 0.1 * 2 * np.pi
x0 = 1
y0 = 0
vx0 = 0
vy0 = 1

#start time = 0
tstart = 0
```

```python
while (tstart<=f):
    xx = x0 + vx0 * dt
    yy = y0 + vy0 * dt
    r = np.sqrt(xx**2 + yy**2)
    vx0 = vx0 - dt*(1*1*xx)/r**3
    vy0 = vy0 - dt*(1*1*yy)/r**3
    x  = append(x,xx)
    y  = append(y,yy)
    vx = append(vx,vx0)
    vy = append(vy,vy0)
    t1 = append(t1,tstart)
    x0 = xx
    y0 = yy
    tstart = tstart + dt
```

```python
#Plotting orbit using Euler method

plt.figure(figsize=(8,8))
plt.plot(x,y,label='Orbit',c='r')
plt.xlabel("x values")
plt.ylabel("y values")
title("The Euler Method")
plt.grid()
plt.legend()
```

The orbit plotted should supposedly look circular; however, with the time range chosen to plot, it looks like it is not.

```python
#Plotting the angular momentum

angmom = x * vy - y * vx
plt.figure(figsize=(8,8))
plot(t1,angmom,color='r',label='Angular Momentum')
plt.xlabel("Time (sec)")
plt.ylabel("Angular Momentum (kgm^2/sec)")
plt.title("Angular Momentum as a Function of Time")
plt.grid()
plt.legend()
plt.ylim(-1.5,1.5)
```

Because the plot produced a straight line above, that means that the angular momentum as a funcion of time is a constant; therefore, it is conserved.
